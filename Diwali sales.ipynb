```python
# import python libraries

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt # visualizing data
%matplotlib inline
import seaborn as sns
```

```python
# import csv file
df = pd.read_csv('Diwali Sales Data.csv', encoding= 'unicode_escape')
```

```python
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1, inplace=True)
```

```python
df.head(10)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df['Amount']= df['Amount'].astype(int)
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Marriage'})
```

```python
df.describe()
```

```python
df[['Amount']].describe()
```

# EDA

```python
ax = sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Gender', y='Amount', data = sales_gen)
```

```python
ax = sns.countplot(x= 'Age Group', data=df, hue= 'Gender')
for bars in ax.containers:   # this statement gives the exact count above the bars as shown below
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x= 'Age Group', data=df, hue= 'Gender')
```

```python
# Total Amt vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group', y='Amount', data= sales_age)
```

```python
df.columns
```

```python
state_odr = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(8)
sns.barplot(x='State', y='Orders',data=state_odr)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
states_amt= df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(8)
sns.barplot(x='State', y='Amount', data=states_amt)
sns.set(rc={'figure.figsize':(15,5)})
```

```python
ax = sns.countplot(x='Marital_Status', data=df)
sns.set(rc={'figure.figsize':(6,5)})
for bars in ax.containers:   # this statement gives the exact count above the bars as shown below
    ax.bar_label(bars)
```

```python
sales_amt = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=sales_amt)
sns.set(rc={'figure.figsize':(5,5)})
```

```python
df.columns
```

```python
sales_zone = df.groupby(['Zone', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Zone', y='Amount', hue='Gender', data=sales_zone)
```

```python
sales_zone = df.groupby(['Zone', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Zone', y='Orders', hue='Gender', data=sales_zone)
```

```python
prdcatg_odrs = df.groupby(['Product_Category', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Orders',y='Product_Category', hue='Gender', data=prdcatg_odrs)
sns.set(rc={'figure.figsize':(5,6)})
```

```python
prdcatg_amt = df.groupby(['Product_Category', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Amount',y='Product_Category', hue='Gender', data=prdcatg_amt)
sns.set(rc={'figure.figsize':(5,6)})
```

```python
occp_odrs = df.groupby(['Occupation', 'Gender'], as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False)
sns.barplot(x='Orders',y='Occupation', hue='Gender', data=occp_odrs)
sns.set(rc={'figure.figsize':(5,6)})
```

```python
occp_amt = df.groupby(['Occupation', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Amount',y='Occupation', hue='Gender', data=occp_amt)
sns.set(rc={'figure.figsize':(5,6)})
```

```python
ax=sns.countplot(y='Occupation', hue='Gender', data=df)
sns.set(rc={'figure.figsize':(5,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax=sns.countplot(y='Occupation', data=df)
sns.set(rc={'figure.figsize':(5,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

# Conclusions
1. Most of the customers are unmarried females and amount spent by them is also greater than males.
2. Most of the customers are of age group 26-35, followed by 36-45 age grp.
3. Most amount spent is by age group of 26-35 followed by 36-45 age grp.
4. Most of the orders are from UPfollowed by Maharastra and Karnataka,also amount spent by states also follow the same trend.
5. Most of the customers are from IT sector followed by Healthcare and Avaiation and amount spent by different sectors also follow the same trend.
6. Most amount is spent by female customers from central zone followed by southern and western.
7. Most orders come from central zone followed by southern and western.
8. Most orders came for product category Clothing followed by Food and Electronics mostly by females.
9. Most amount was spent on Food followed by Footwear and Clothing mostly by females.
